```python
%matplotlib inline
import os
import matplotlib.pyplot as plt

import torch
import torchvision
import torch.utils.data as data
from torchvision import transforms
```

```python
TRAIN_DATA_PATH = '~/Desktop/hse_studies/signs-classification/data/Train'
TEST_DATA_PATH = '~/Desktop/hse_studies/signs-classification/data/Test'
```

```python
transform_pipeline = transforms.Compose([
    transforms.Resize(32),
    transforms.ToTensor()
])
```

```python
train_data = torchvision.datasets.ImageFolder(root=TRAIN_DATA_PATH, transform=transform_pipeline)
```

```python
train_data_loader = data.DataLoader(train_data, batch_size=2, shuffle=True, num_workers=2)
```

```python
fig = plt.figure(figsize=(8, 8))
col, row = 3, 3
for i in range(1, col * row + 1):
    sample_index = torch.randint(len(train_data), size=(1,)).item()
    img, label = train_data[sample_index]
    print(img.size())
    fig.add_subplot(row, col, i)
    plt.title(label)
    plt.axis("off")
    plt.imshow(img.permute(1, 2, 0), cmap="gray")
plt.show()
```
